```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample data
cities = ['new york', 'mumbai', 'paris']
max_temperatures = [36, 92, 54]

# Creating a bar graph
plt.figure(figsize=(8, 6))
plt.bar(cities, max_temperatures)
plt.xlabel('Cities')
plt.ylabel('Max Temperature (Â°C)')
plt.title('Max Temperature in Different Cities')
plt.show()

# Creating a pie chart
plt.figure(figsize=(8, 6))
plt.pie(max_temperatures, labels=cities, autopct='%1.1f%%')
plt.title('Max Temperature Distribution in Different Cities')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
df1 = pd.read_csv('/content/company_sales_data.csv')
months = df1['month_number']
profits = df1['total_profit']



# Creating a line graph
plt.figure(figsize=(8, 6))
plt.plot(months, profits, marker='o', linestyle='-', color='b')
plt.xlabel('Months')
plt.ylabel('Profit ($)')
plt.title('Profit for Each Month')
plt.grid(True)
plt.show()
```

```python
product_columns = df1.columns[1:7]
product_sales = [df1[col].sum() for col in product_columns]

# Creating a pie chart
plt.figure(figsize=(8, 6))
plt.pie(product_sales, labels=product_columns, autopct='%1.1f%%')
plt.title('Sales Data for All Products')
plt.show()
```

```python

# Extract the sales data for face cream and facewash
days = df1['month_number']
facewash_sales = df1['facewash']

# Creating a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(days, facewash_sales, label='Facewash')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Face Cream and Facewash Sales Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
toothpaste = df1['toothpaste']
facecream = df1['facecream']

# Creating a bar graph
plt.figure(figsize=(8, 6))
plt.bar(toothpaste, facecream)
plt.xlabel('TP')
plt.ylabel('FC')
plt.title('TP vs FC')
plt.show()
```
