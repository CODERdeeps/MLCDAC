```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Generate x values from -π to +π
x = np.linspace(-np.pi, np.pi, 100)

# Compute sin and cosine values for each x
sin_values = np.sin(x)
cos_values = np.cos(x)

# Plot sin and cosine waves
plt.plot(x, sin_values, label='sin(x)')
plt.plot(x, cos_values, label='cos(x)')

# Set plot labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin and Cosine Waves')

# Display legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Create the list
data = [12, 14, 11, 18, 20, 23]

# Generate x values as indices of the data points
x = range(len(data))

# Plot the line graph
plt.plot(x, data, marker='o')

# Set plot labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Graph')

# Display the plot
plt.show()
```

```python
# Create the list
data = [12, 14, 11, 18, 20, 23]

# Create labels for each value
labels = ['A', 'B', 'C', 'D', 'E', 'F']

# Plot the pie plot
plt.pie(data, labels=labels, autopct='%1.1f%%')

# Set the title of the plot
plt.title('Pie Plot')

# Display the plot
plt.show()




```

```python
# Create the dataframe
data = {'prod': [31, 24, 26, 12, 19],
        'yr': [1992, 1999, 1998, 1996, 1993]}
index = ['pd', 'sc', 'mz', 'ln', 'bl']
food = pd.DataFrame(data, index=index)

# Extract the data from the row labeled 'sc'
sc_data = food.loc['sc']

# Display the production for 'sc' and 'pd' rows
production_sc_pd = food.loc[['sc', 'pd'], 'prod']
print(production_sc_pd)

# Create bar chart for the dataframe
plt.bar(food['yr'], food['prod'])
plt.xlabel('Year')
plt.ylabel('Production')
plt.title('Production by Year')
plt.show()


```
